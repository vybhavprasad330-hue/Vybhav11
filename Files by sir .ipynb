## What is a File?

A **file** is a collection of data stored permanently on secondary storage (hard disk).

### Why Files?

* Data persistence
* Large data storage
* Sharing data between programs

### Types of Files

* **Text files**: '.txt', '.csv', '.log'
* **Binary files**: '.jpg', '.pdf', '.exe'

## Opening a File in Python

file_object = open("filename", "mode")

### File Modes

| Mode | Meaning           |
| ---- | ----------------- |
| 'r'  |    Read              |
| 'w'  |    Write (overwrite) |
| 'a'  |    Append            |
| 'r+' |    Read + Write      |

# read() reads the entire file at once

*Assume crop text file contains*

 - Rice
 - Wheat
 - Maize
 - Pulses

Reads everything in the file as one single string.

content = file.read()

Output:

 - Rice
 - Wheat
 - Maize
 - Pulses

Internally:

"Rice\nWheat\nMaize\nPulses\n"

 - Suitable for small files
 - Not memory-efficient for large datasets


```python
### readline()
#Reads only the first line (up to \n).
#Useful when reading line-by-line
#Memory-efficient

file = open("crop1.txt", "r")
print(file.readline())
print(file.readline())
#print(file.readline())
#print(file.readline())
#print(file.readline())
file.close()
```

```python
# no need to call close()
with open("crop1.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
#Reads entire file at once
#Stores all lines in a list

with open("crop.txt", "r") as file:
    lines = file.readlines()

#['Rice\n', 'Wheat\n', 'Maize\n', 'Pulses\n']

for line in lines:
    print(line.strip())
```

```python
### Problem Without 'with'

#* Forgetting to close files
#* Resource leakage

### Solution: `with`, File closes automatically

with open("crop.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
#Using split() with File Handling
with open("crop_data.txt", "r") as file:
    for line in file:
        fields = line.strip().split(",")
        print(fields)

```

```python
#Accessing Split Data

with open("crop_data.txt", "r") as file:
    for line in file:
        fields = line.strip().split(",")

        crop_code = fields[0]
        crop_name = fields[1]
        season = fields[2]
        yield_level = fields[3]

        print(crop_code, crop_name, season, yield_level)
    print(type(line.split(",")))
```
