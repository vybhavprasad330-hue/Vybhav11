```python
crops = 0
points = 0

while True:
    print("🌾 Welcome to Farm Game 🌾")

    print("1. Plant seed")
    print("2. Water crops")
    print("3. Harvest crops")
    print("4. Exit")

    choice = input("Choose an action: ")

    if choice == "1":
        crops += 1
        print("🌱 You planted a seed.")

    elif choice == "2":
        if crops > 0:
            print("💧 Crops watered.")
        else:
            print("No crops to water!")

    elif choice == "3":
        if crops > 0:
            points += crops * 10
            print("🌾 You harvested", crops, "crops!")
            crops = 0
        else:
            print("Nothing to harvest!")

    elif choice == "4":
        print("Game Over! Total points:", points)
        break

    else:
        print("Invalid choice!")
```

```python
import random

# Initial values
money = 100
wheat = 0
rice = 0
day = 1

print("🌾 WELCOME TO ADVANCED FARM GAME 🌾")

while True:
    print("\n📅 Day:", day)
    print("💰 Money:", money)
    print("🌱 Wheat:", wheat, "| 🌾 Rice:", rice)

    weather = random.choice(["Sunny", "Rainy", "Storm"])
    print("⛅ Weather today:", weather)

    print("\n1. Buy Wheat Seeds (₹10 each)")
    print("2. Buy Rice Seeds (₹15 each)")
    print("3. Harvest Crops")
    print("4. Skip Day")
    print("5. Exit Game")

    choice = input("Choose an action: ")

    if choice == "1":
        if money >= 10:
            wheat += 1
            money -= 10
            print("🌱 Wheat seed planted.")
        else:
            print("❌ Not enough money.")

    elif choice == "2":
        if money >= 15:
            rice += 1
            money -= 15
            print("🌾 Rice seed planted.")
        else:
            print("❌ Not enough money.")

    elif choice == "3":
        if wheat > 0 or rice > 0:
            earnings = wheat * 20 + rice * 30
            money += earnings
            print("✅ Crops harvested! Earned ₹", earnings)
            wheat = 0
            rice = 0
        else:
            print("❌ No crops to harvest.")

    elif choice == "4":
        print("⏩ Day skipped.")

    elif choice == "5":
        print("👋 Game Over! Final Money:", money)
        break

    else:
        print("⚠️ Invalid choice.")

    # Weather impact
    if weather == "Storm":
        wheat = max(0, wheat - 1)
        rice = max(0, rice - 1)
        print("🌪️ Storm damaged some crops!")

    day += 1

    # Win/Lose Conditions
    if money >= 500:
        print("🏆 CONGRATULATIONS! You became a successful farmer!")
        break

    if money <= 0 and wheat == 0 and rice == 0:
        print(" You ran out of money and crops. Game Over!")
        break
```

```python
# Initial inputs
money = int(input("Enter initial money: "))
day = 1
wheat = 0
rice = 0

print("\n🌾 FARMING GAME STARTED 🌾")

while True:
    print("\n📅 Day:", day)
    print("💰 Money:", money)
    print("🌱 Wheat:", wheat, "| 🌾 Rice:", rice)

    # User-defined prices
    wheat_seed_price = int(input("\nEnter Wheat seed price: "))
    rice_seed_price = int(input("Enter Rice seed price: "))
    wheat_sell_price = int(input("Enter Wheat selling price: "))
    rice_sell_price = int(input("Enter Rice selling price: "))

    print("\n1. Buy Wheat Seeds")
    print("2. Buy Rice Seeds")
    print("3. Harvest Crops")
    print("4. Skip Day")
    print("5. Exit Game")

    choice = input("Choose an action: ")

    if choice == "1":
        qty = int(input("How many Wheat seeds to buy? "))
        cost = qty * wheat_seed_price
        if money >= cost:
            wheat += qty
            money -= cost
            print("🌱 Wheat seeds planted:", qty)
        else:
            print("❌ Not enough money.")

    elif choice == "2":
        qty = int(input("How many Rice seeds to buy? "))
        cost = qty * rice_seed_price
        if money >= cost:
            rice += qty
            money -= cost
            print("🌾 Rice seeds planted:", qty)
        else:
            print("❌ Not enough money.")

    elif choice == "3":
        if wheat > 0 or rice > 0:
            earnings = (wheat * wheat_sell_price) + (rice * rice_sell_price)
            money += earnings
            print("✅ Harvest successful! Earned ₹", earnings)
            wheat = 0
            rice = 0
        else:
            print("❌ No crops to harvest.")

    elif choice == "4":
        weather = input("Enter weather (Sunny/Rainy/Storm): ")

        if weather.lower() == "storm":
            wheat_loss = int(input("Enter Wheat loss due to storm: "))
            rice_loss = int(input("Enter Rice loss due to storm: "))
            wheat = max(0, wheat - wheat_loss)
            rice = max(0, rice - rice_loss)
            print("🌪️ Crops damaged due to storm.")
        else:
            print("⛅ No crop damage today.")

    elif choice == "5":
        print("👋 Game Over!")
        print("Final Money:", money)
        break

    else:
        print("⚠️ Invalid choice.")

    day += 1

    # Win / Lose conditions (input-based)
    target = int(input("Enter target money to win: "))
    if money >= target:
        print("🏆 YOU WON! Target money achieved.")
        break

    if money <= 0 and wheat == 0 and rice == 0:
        print(" You lost! No money and no crops.")
        break
```
