**Dictionary**

Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

clear() - Removes all the elements from the dictionary

copy() - Returns a copy of the dictionary

fromkeys() - Returns a dictionary with the specified keys and value

get() - Returns the value of the specified key

items() - -Returns a list containing a tuple for each key value pair

keys()	Returns a list containing the dictionary's keys

pop() - Removes the element with the specified key

popitem() - Removes the last inserted key-value pair

setdefault() -

Returns the value of the specified key.
If the key does not exist: insert the key, with the specified value

update() - Updates the dictionary with the specified key-value pairs

values() - Returns a list of all the values in the dictionary

**Access Dictionary Items**

```python
thisdict = {
    "crop": "Rice",
    "variety": "IR-64",
    "year": 2022
}

# Access value using key
x = thisdict["variety"]
print(x)


# get() method

x = thisdict.get("variety")
print(x)


# Get Keys
# The keys() method returns a view of all keys
#show the usage of list and tuple constructor here

x = thisdict.keys()
print(x)


# Add a new item and observe keys update
farm = {
    "crop": "Rice",
    "variety": "IR-64",
    "year": 2022
}

x = farm.keys()
print(x)   # before the change

farm["season"] = "Kharif"

print(x)   # after the change
```

**Get Values**

```python
# The values() method returns a view of all values

thisdict = {
    "crop": "Rice",
    "variety": "IR-64",
    "year": 2022
}

x = thisdict.values()
print(x)

# Modify original dictionary and see values update
farm = {
    "crop": "Rice",
    "variety": "IR-64",
    "year": 2022
}

x = farm.values()
print(x)   # before the change

farm["year"] = 2024

print(x)   # after the change
```

```python
#Example

# Original dictionary
student = {
    "name": "Anita",
    "roll_no": 101,
    "branch": "CSE",
    "marks": 88
}

print("Original dictionary:", student)

#copy() — Creates a shallow copy

student_copy = student.copy()
print("Copied dictionary:", student_copy)

#pop() — Removes a specific key

removed_value = student.pop("marks")
print("Removed value:", removed_value)
print("After pop():", student)

#popitem() — Removes the last inserted item

last_item = student.popitem()
print("Removed item:", last_item)
print("After popitem():", student)

#update() — Adds or modifies key–value pairs

student.update({"year": 2, "branch": "AI & DS"})
print("After update():", student)

#clear() — Removes all items, but student dictionary will still be available,
#but it will be empty

student.clear()
print("After clear():", student)

# del function will delete the dictionary altogether
del student
#print("After deletion",student)
```

Enumerate()

Scenario

You have a list of crops grown in different plots, and you want to:

*   Print the plot number
*   Along with the crop name

crops = ["Rice", "Wheat", "Maize", "Cotton"]




```python
#Traditional way

crops = ["Rice", "Wheat", "Maize", "Cotton"]

i = 0
for crop in crops:
    print("Plot", i, ":", crop)
    i += 1
```

Problems here

  - Extra variable i
  - Risk of forgetting i += 1
  - Code looks less clean
  - Easy to make mistakes

```python
crops = ["Rice", "Wheat", "Maize", "Cotton"]

for plot_no, crop in enumerate(crops):
    print("Plot", plot_no, ":", crop)
```

```python
crops = ["Rice", "Wheat", "Maize", "Cotton"]

for crop in enumerate(crops):
    print("Plot", plot_no, ":", crop)
```

What enumerate() Actually Returns

*[(0, 'Rice'), (1, 'Wheat'), (2, 'Maize'), (3, 'Cotton')]*

Each item is a tuple:

*(index,value)*

for index, value in enumerate(...)

```python
yields = [1200, 1500, 1100, 900]

for plot_no, yield_kg in enumerate(yields, start=1):
    print(f"Plot {plot_no} yield: {yield_kg} kg")
```

**Zip() Function**

Iterating over multiple related sequences at the same time

You have:

 - Plot numbers
 - Crop names
 - Yield values

All belong together — row-wise

 *plots  = [1, 2, 3, 4]
crops  = ["Rice", "Wheat", "Maize", "Cotton"]
yields = [1200, 1500, 1100, 900]*

You want output like:

Plot 1 → Rice → 1200 kg


```python
#Without zip()

plots = [1, 2, 3, 4]
crops = ["Rice", "Wheat", "Maize", "Cotton"]
yields = [1200, 1500, 1100, 900]

for i in range(len(plots)):
    print(plots[i], crops[i], yields[i])

# Depends on index
# Risky if lengths differ
```

```python
plots = [1, 2, 3, 4]
crops = ["Rice", "Wheat", "Maize", "Cotton"]
yields = [1200, 1500, 1100]

for i in range(len(plots)):
    print(plots[i], crops[i], yields[i])
```

```python
#With zip()

plots = [1, 2, 3, 4]
crops = ["Rice", "Wheat", "Maize", "Cotton"]
yields = [1200, 1500, 1100, 900]

for plot, crop, yield_kg in zip(plots, crops, yields):
    print(f"Plot {plot} grows {crop} with yield {yield_kg} kg")
```

```python
plots = [1, 2, 3, 4]
crops = ["Rice", "Wheat", "Maize", "Cotton"]
yields = [1200, 1500, 1100]

for plot, crop, yield_kg in zip(plots, crops, yields):
    print(f"Plot {plot} grows {crop} with yield {yield_kg} kg")
```

**What does zip() actually return?**

list(zip(plots, crops, yields))

Output:

[(1, 'Rice', 1200), (2, 'Wheat', 1500),
 (3, 'Maize', 1100), (4, 'Cotton', 900)]

 - Each element is a tuple
 - Elements are grouped position-wise

```python
#Combining enumerate() and zip()

crops  = ["Rice", "Wheat", "Maize"]
yields = [1200, 1500, 1100]

for plot_no, (crop, yield_kg) in enumerate(zip(crops, yields), start=1):
    print(f"Plot {plot_no}: {crop} → {yield_kg} kg")
```

```python
# unzipping the records

records = [
    (1, "Rice", 1200),
    (2, "Wheat", 1500),
    (3, "Maize", 1100)
]

plot_no, crop, yield_kg = zip(*records)

print(plot_no)
print(crop)
print(yield_kg)
```

```python
#Zip two dictionaries (DEFAULT behavior)
#By default, zip() iterates over dictionary keys

crop_price = {
    "Rice": 40,
    "Wheat": 35,
    "Maize": 30
}

crop_yield = {
    "Rice": 1200,
    "Wheat": 1500,
    "Maize": 1100
}

for a, b in zip(crop_price, crop_yield):
    print(a, b)
```

```python
#Zip dictionary KEYS with VALUES

for crop, price in zip(crop_price.keys(), crop_price.values()):
    print(crop, price)
```

```python
for crop, price, yield_kg in zip(
        crop_price.keys(),
        crop_price.values(),
        crop_yield.values()):
    print(f"{crop}: Price={price}, Yield={yield_kg}")
```

```python
#Zip + Unzip with Dictionary Data

items = list(crop_price.items())
keys, values = zip(*items)

print(keys)
print(values)
```
