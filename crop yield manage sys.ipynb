```python
# Crop Yield Management System

# Function to display all recorded crop yields
def display_yields(yields):
    print("""Recorded crop yields (kg):""")
    for i in range(len(yields)):
        print(f"Field {i + 1}: {yields[i]} kg")

# Function to calculate and return the average crop yield
def calculate_average(yields):
    if len(yields) == 0:
        return 0
    total = 0
    for value in yields:
        total += value
    average = total / len(yields)
    return average

# Function to identify and return all crop yields above a given threshold
def yields_above_threshold(yields, threshold):
    above = []
    for value in yields:
        if value > threshold:
            above.append(value)
    return above


# Main program
def main():
    # Accept number of fields
    n = int(input("Enter the number of fields: "))

    yields = []

    # Collect crop yield for each field
    for i in range(n):
        value = float(input(f"Enter yield (kg) for field {i + 1}: "))
        yields.append(value)

    # Display all recorded yields
    display_yields(yields)

    # Calculate statistics
    max_yield = max(yields)
    min_yield = min(yields)
    avg_yield = calculate_average(yields)

    # Ask for threshold and get yields above threshold
    threshold = float(input("""Enter threshold yield (kg): """))
    above_list = yields_above_threshold(yields, threshold)

    # Display required outputs
    print("""----- Crop Yield Summary -----""")
    print(f"Total number of fields: {len(yields)}")
    print(f"Maximum yield: {max_yield} kg")
    print(f"Minimum yield: {min_yield} kg")
    print(f"Average yield: {avg_yield:.2f} kg")

    print("""Yields above threshold:""")
    if len(above_list) == 0:
        print("No yields above the given threshold.")
    else:
        for value in above_list:
            print(f"{value} kg")


# Call main function
if __name__ == "__main__":
    main()
```
