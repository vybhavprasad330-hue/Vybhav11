```python
#No argument is passed
def number_of_crops():
    # Let's assume we have a fixed number of crops in the field.
    return 150  # Example: 150 crops
number_of_crops()
```

```python
#No argument is passed, no value is returned

def weather_update():
    print("The weather today is sunny with no rain expected.")
weather_update()
```

```python
#Only one argument is passed, no value is returned also called as
#void functions

def plant_info(plant_name):
    print(f"The {plant_name}, requires at least 8 hours of sunlight per day.")
plant_info("snake tree")
```

```python
#Only one argument is passed, value is returned

def water_needs(plant_type):
    if plant_type == "corn":
        return 30  # Corn needs 30 liters of water per day
    elif plant_type == "wheat":
        return 20  # Wheat needs 20 liters of water per day
    else:
        return 25  # Default water needs
water_needs("wheat")
```

```python
#Two arguments are passed, no value is returned

def irrigation_needed(moisture_level, rainfall):
    if moisture_level <= 40 and rainfall <= 10:
        print("Irrigation needed.")
    else:
        print("Irrigation not needed.")
irrigation_needed(40, 10)
```

```python
#Function where a boolean value is returned

def is_ready_for_harvest(days_since_planted,plant_type):
    if days_since_planted >= 120 and plant_type=="Rice":
        return True
    else:
        return False
plant_type=input("Enter the plant type")
days_since_planted=int(input("Enter the number of days since planted"))
is_ready_for_harvest(days_since_planted,plant_type)
```

```python
#Returning Multiple Values

def weather_conditions():
    temperature = 28  # in Celsius
    humidity = 70  # in percentage
    return temperature, humidity  # returns a tuple

# Usage
temp, hum =weather_conditions()
print(f"Temperature: {temp}C, Humidity: {hum}%")
```

```python
#Returning Collections (Lists, Dictionaries, etc.)

def crops_in_field():
    return ["Corn", "Wheat", "Rice"]

# Usage
field_crops = crops_in_field()
print(field_crops)  # ['Corn', 'Wheat', 'Rice']
```

**Special Types of Arguments**

```python
#Default Arguments
#You can set default values for arguments.
#If no argument is passed, the default value will be used.

def water_needs(plant_type, water_amount=20):
    print(f"{plant_type} needs {water_amount} liters of water.")

# Usage
water_needs("Corn")  # Uses default value of 20
water_needs("Rice", 15)  # Uses provided value of 15
```

```python
#Keyword Arguments

def crop_info(crop_name, yield_per_hectare, growth_rate):
    print(f"Crop: {crop_name}, Yield: {yield_per_hectare} tons/ha, Growth Rate: {growth_rate}")

# Usage
crop_info(crop_name="Wheat", yield_per_hectare=25, growth_rate="Medium")
```

```python
#Unpacking Arguments

def irrigation(*water_needs):
    for water in water_needs:
        print(f"Water needed: {water} liters")
        list1=list(water_needs)
        list1[0]=0
        print(list1)

crops_water = [25, 20, 15]
irrigation(*crops_water)  # Unpacks the list and passes the values
```

```python
nction for addition
def add(x, y):
    return x + y

# Function for subtraction
def subtract(x, y):
    return x - y

# Function for multiplication
def multiply(x, y):
    return x * y

# Function for division
def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

# Function to take input values from the user
def get_input():
    try:
        # Take user input for the first number
        x = float(input("Enter the first number: "))
        # Take user input for the second number
        y = float(input("Enter the second number: "))
        return x, y

    except ValueError:
        print("Invalid input! Please enter valid numbers.")
        return get_input()  # Recursive call to get valid input

# Function to print the menu of operations
def print_menu():
    print("\nSelect operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

# Main calculator function
def calculator():
    while True:
        print_menu()

        # Get the user's choice
        choice = input("Enter choice (1/2/3/4/5): ")

        if choice == '5':
            print("Exiting the calculator.")
            break  # Exit the program

        # Call get_input function to get the values
        x, y = get_input()

        # Perform operation based on user choice
        if choice == '1':
            print(f"{x} + {y} = {add(x, y)}")
        elif choice == '2':
            print(f"{x} - {y} = {subtract(x, y)}")
        elif choice == '3':
            print(f"{x} * {y} = {multiply(x, y)}")
        elif choice == '4':
            print(f"{x} / {y} = {divide(x, y)}")
        else:
            print("Invalid Input! Please select a valid operation (1/2/3/4/5).")

# Directly call the calculator function
calculator()
print("done with the program")
```
