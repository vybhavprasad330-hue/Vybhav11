```python
#local function
def myfunc():
  x = 300
  print(x)

myfunc()
```

```python
#A variable created outside of a function is global and can be used by anyone:
x = 300
def myfunc():
  print(x)

myfunc()

print(x)
```

```python
x = 300

def myfunc():
  x = 200
  print(x)

myfunc()

print(x)
```

```python
#If you use the global keyword, the variable belongs to the global scope:

def myfunc():
  global x
  x = 300

myfunc()

print(x)
```

```python
#lambda function
x = lambda a : a + 10
print(x(5))
```

```python
#2 argumrnts
x = lambda a, b : a * b
print(x(5, 6))
```

```python
#3 arguments
x = lambda a, b, c : a + b + c
print(x(5, 6, 2))
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)

print(mydoubler(11))
```

```python
def myfunc(n):
  return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(11))
print(mytripler(11))
```

```python
def myfunc(n,p):
  return lambda a,b : a * n + p*b

mydoubler = myfunc(2,3)

print(mydoubler(11,22))
```

```python
def myfunc(n,p,q):
  return lambda a,b : a * n + p * b / q

mylambda = myfunc(2,3,4)

print(mylambda(11,22))
```

```python
#storting sting by length
words = ["apple", "pie", "banana", "cherry"]
sorted_words = sorted(words, key=lambda x: len(x))
print(sorted_words)
```

```python
students = [("Emil", 25), ("Tobias", 22), ("Linus", 28)]
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
```
