```python
# crop suitability check:
soil_moisture=float(input("Enter soil moisture in % : "))
rainfall=float(input("Enter rainfall in mm : "))

if soil_moisture >= 70 and rainfall > 500 :
  print("Suitable for rice cultivation")

else:
  print("Not suitable for rice cultivation")
```

```python
# to check fertilizer requirement
crop=input("Enter crop name (rice/wheat/maize/tomato) : ")
N=float(input("Enter nitrogen content in kg/ha : "))
p=float(input("Enter phosphorus content in kg/ha : "))
k=float(input("enter potassium content in kg/ha : "))

if crop in ['rice', 'wheat', 'maize', 'tomato'] and (N <= 140 or p <= 100 or k <= 100) :
  print("Fertilizer requirement is not met")

else:
  print("Fertilizer requirement is met")
```

```python
# Average Rainfall Analysis
rainfall = []

for i in range(5):
    value = float(input(f"Enter rainfall for day {i+1}: "))
    rainfall.append(value)

average = sum(rainfall) / 5

print("Average rainfall:", average)

if average >= 100:
    print("Sufficient rainfall")
else:
    print("Insufficient rainfall")
```
