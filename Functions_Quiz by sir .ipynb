```python
#Which output occurs when calling process_tuple((10, 20, 30))?

def process_tuple(tup):
    total = 0
    i = 0
    while i < len(tup):
        if tup[i] % 2 == 0:
            total += tup[i]
        i += 1
    return total

# Call the function and print its return value
print(process_tuple((10, 20, 30)))
```

```python
#What is the output of test_control((1,2,3,4))?

def test_control(tup):
    count_even = 0
    for val in tup:
        if val % 2 == 0:
            count_even += 1
        elif val > 2:
            continue
    return count_even
test_control((1,2,3,4))
```

```python
#What will  print(modify_inplace([10,20,30])) display?

def modify_inplace(nums):
    i = 0
    while i < len(nums):
        if nums[i] < 25:
            nums[i] *= 2
        else:
            del nums[i]
            continue
        i += 1
    return nums
print(modify_inplace([10,20,30]))
```

```python
#What does nest_func([[1,2],[3]]) print?

def nest_func(nested):
    for sublist in nested:
        if len(sublist) > 1:
            print(sum(sublist))
nest_func([[1,2],[3]])
```

```python
#What is the output of the following code?

def analyze(lists):
    empty_count = 0
    i = 0
    while i < len(lists):
        if not lists[i]:
            empty_count += 1
        i += 1
    return empty_count
print(analyze([[], [1], [2,3]]))
```

```python
#What does this function return when called as mystery([1, 2, 2, 3])?

def mystery(lst):
    seen = []
    for x in lst:
        if x not in seen:
            seen.append(x)
    return seen
print(mystery([1, 2, 2, 3]))
```

```python
#Consider this code using nested loops and a condition:

nums = [1, 2, 3]
output = []

for i in range(len(nums)):
    for j in range(len(nums)):
        if i == j:
            continue
        if nums[i] + nums[j] > 4:
            output.append((nums[i], nums[j]))

print(output)

#What is printed?
```

```python
#What is the result of this code with nested loops and break?

pairs = []

for i in range(3):
    for j in range(3):
        if j == 1:
            break
        pairs.append((i, j))

print(pairs)
```

```python
crops = [25, 18, 35]  # yields in tons/hectare
pairs = []

for i in range(len(crops)):
    for j in range(i+1, len(crops)):
        if crops[i] + crops[j] > 50:
            pairs.append((crops[i], crops[j], "High"))
        elif crops[i] + crops[j] > 40:
            pairs.append((crops[i], crops[j], "Medium"))
print(pairs)


#What is printed?
```

```python
fields = [[10, 0, 15], [0, 20, 0], [5, 10, 25]]  # soil moisture
irrigated = 0

for row in range(len(fields)):
    for col in range(len(fields[row])):
        if fields[row][col] < 12:
            irrigated += 1
            continue
        if fields[row][col] > 20:
            break
print(irrigated)

#What is printed?
```

```python
costs = [30, 45, 25]  # Rs/kg for N,P,K
rates = [2, 1.5, 3]

recommendations = []
for i in range(len(costs)):
    for j in range(len(rates)):
        if i != j and costs[i] * rates[j] < 80:
            recommendations.append(f"F{i+1}K{j+1}")
print(recommendations)

#what is printed?
```

```python
scans = [[1,0,1], [0,1,0], [1,0,0]]  # pest presence
alerts = []
for i in range(len(scans)):
    row_count = 0
    for j in range(len(scans[i])):
        if scans[i][j] == 1:
            row_count += 1
    if row_count > 1:
        alerts.append(f"Row{i} quarantine")
print(alerts)

#What is printed?
```

```python
#  Harvest Window Calculation

days = [5,3,7,2]  # days to harvest
windows = []
i = 0
while i < len(days)-1:
    j = i + 1
    while j < len(days):
        if days[i] + days[j] <= 8:
            windows.append((i,j))
        j += 1
    i += 1
print(windows)

#What prints?
```
