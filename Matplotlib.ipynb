```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# FALL OF WICKETS DATA
# (Overs vs Runs at that wicket)
# -----------------------------

# India (approx progression)
ind_overs = [0.4, 1.2, 3.5, 6.1, 9.4, 12.3, 15.0, 18.5]
ind_runs  = [5, 12, 28, 45, 60, 78, 95, 111]

# South Africa
sa_overs = [0.5, 1.3, 3.2, 9.0, 14.5, 17.2, 19.4]
sa_runs  = [8, 15, 32, 85, 130, 165, 187]

# -----------------------------
# STYLE SETTINGS
# -----------------------------
plt.figure(figsize=(12,7))
# plt.style.use('white') # Removed the problematic line

# Plot lines
plt.plot(ind_overs, ind_runs, marker='o', linewidth=3, label='India')
plt.plot(sa_overs, sa_runs, marker='s', linewidth=3, label='South Africa')

# Fill area under curves
plt.fill_between(ind_overs, ind_runs, alpha=0.15)
plt.fill_between(sa_overs, sa_runs, alpha=0.15)

# Labels and title
plt.xlabel("Overs", fontsize=12)
plt.ylabel("Runs at Fall of Wicket", fontsize=12)
plt.title("🏏 IND vs SA — Fall of Wickets (Runs Comparison)", fontsize=16, weight='bold')

# Grid styling
plt.grid(alpha=0.25, linestyle='--')

# Legend
plt.legend(frameon=False, fontsize=11)

# Annotate final scores
plt.annotate("IND All Out",
             (ind_overs[-1], ind_runs[-1]),
             textcoords="offset points",
             xytext=(8,8))

plt.annotate("SA Innings End",
             (sa_overs[-1], sa_runs[-1]),
             textcoords="offset points",
             xytext=(8,8))

plt.xlabel("Overs")
plt.ylabel("Score")
plt.title("Fall of Wickets Comparison - IND vs SA (T20 World Cup)")
plt.tight_layout()
plt.show()
```

Smart Agriculture Monitoring System

A smart farming system monitors different parameters across multiple farm plots.
The dashboard displays:
🌱 Plot 1 – Soil Moisture
🌾 Plot 2 – Crop Yield
🌡 Plot 3 – Temperature
💧 Plot 4 – Rainfall
🧪 Plot 5 – Soil pH
🌞 Plot 6 – Sunlight Hours
🌽 Plot 7 – Fertilizer Usage
The farmer wants all data visualized in one dashboard using subplots.

```python
import matplotlib.pyplot as plt
import numpy as np

days = np.array([1, 2, 3, 4])

# 1. Soil Moisture (%)
plt.subplot(3, 3, 1)
plt.plot(days, [30, 45, 35, 50])
plt.title("Plot 1 - Soil Moisture")

# 2. Crop Yield (tons)
plt.subplot(3, 3, 2)
plt.plot(days, [2, 3, 4, 5])
plt.title("Plot 2 - Crop Yield")

# 3. Temperature (°C)
plt.subplot(3, 3, 3)
plt.plot(days, [28, 30, 29, 31])
plt.title("Plot 3 - Temperature")

# 4. Rainfall (mm)
plt.subplot(3, 3, 4)
plt.plot(days, [5, 10, 2, 8])
plt.title("Plot 4 - Rainfall")

# 5. Soil pH
plt.subplot(3, 3, 5)
plt.plot(days, [6.5, 6.8, 6.7, 6.9])
plt.title("Plot 5 - Soil pH")

# 6. Sunlight Hours
plt.subplot(3, 3, 6)
plt.plot(days, [8, 9, 7, 10])
plt.title("Plot 6 - Sunlight")

# 7. Fertilizer Usage (kg)
plt.subplot(3, 3, 7)
plt.plot(days, [20, 25, 22, 30])
plt.title("Plot 7 - Fertilizer Usage")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Agricultural data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = np.array([45, 52, 68, 120, 180, 220, 195, 160, 110, 75, 58, 40])
temperature = np.array([18, 20, 24, 28, 32, 35, 34, 33, 30, 26, 22, 19])
wheat_yield = np.array([2.1, 2.3, 2.8, 4.2, 5.8, 6.5, 5.9, 5.2, 3.8, 2.9, 2.4, 2.0])

# Example 1: Line Plot - Rainfall Over Time
plt.figure(figsize=(10, 5))
plt.plot(months, rainfall, marker='o', linewidth=2,
         color='#1f77b4', label='Rainfall')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Rainfall (mm)', fontsize=12)
plt.title('Monthly Rainfall Pattern', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
# plt.savefig('rainfall_pattern.png', dpi=300)
print('Figure 1: Line plot created - Monthly Rainfall Pattern')

# Example 2: Multiple Lines - Temperature and Rainfall
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(months, rainfall, marker='s', color='blue', label='Rainfall')
ax1.set_xlabel('Month', fontsize=12)
ax1.set_ylabel('Rainfall (mm)', color='blue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(months, temperature, marker='^', color='red', label='Temperature')
ax2.set_ylabel('Temperature (°C)', color='red', fontsize=12)
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Rainfall and Temperature Trends', fontsize=14, fontweight='bold')
fig.tight_layout()
print('Figure 2: Dual-axis plot created - Rainfall and Temperature')

# Example 3: Bar Chart - Crop Yields by Region
regions = ['North', 'South', 'East', 'West', 'Central']
yields = [180, 228, 416, 164, 245]
colors_list = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']

plt.figure(figsize=(10, 6))
bars = plt.bar(regions, yields, color=colors_list, edgecolor='black', linewidth=1.5)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Yield (tons)', fontsize=12)
plt.title('Crop Yield by Region', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}',
             ha='center', va='bottom', fontsize=10, fontweight='bold')

print('Figure 3: Bar chart created - Crop Yield by Region')

# Example 4: Scatter Plot - Rainfall vs Yield Correlation
plt.figure(figsize=(10, 6))
plt.scatter(rainfall, wheat_yield, s=100, c=temperature,
            cmap='YlOrRd', edgecolors='black', linewidth=1, alpha=0.7)
plt.colorbar(label='Temperature (°C)')
plt.xlabel('Rainfall (mm)', fontsize=12)
plt.ylabel('Wheat Yield (tons/ha)', fontsize=12)
plt.title('Rainfall vs Yield Correlation (colored by temperature)',
          fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
print('Figure 4: Scatter plot created - Rainfall vs Yield')

# Example 5: Histogram - Yield Distribution
plt.figure(figsize=(10, 6))
plt.hist(wheat_yield, bins=8, color='green', alpha=0.7,
         edgecolor='black', linewidth=1.5)
plt.xlabel('Yield (tons/ha)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Wheat Yields', fontsize=14, fontweight='bold')
plt.axvline(wheat_yield.mean(), color='red', linestyle='--',
            linewidth=2, label=f'Mean: {wheat_yield.mean():.2f}')
plt.legend()
plt.grid(axis='y', alpha=0.3)
print('Figure 5: Histogram created - Yield Distribution')

print('\nAll plots generated successfully!')
print('Note: Use plt.show() to display plots interactively')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- Simulate Data for Runs Scored Per Over ---
num_overs = 20
num_matches = 10
np.random.seed(42) # for reproducibility

# Create a dictionary to hold runs per over data for box plots
# Each over will have 'num_matches' data points
runs_data = {}
for over in range(1, num_overs + 1):
    # Simulate runs for each over with some variability
    # Powerplay overs (1-6), Middle overs (7-15), Death overs (16-20)
    if over <= 6:
        runs = np.random.normal(loc=8 + over*0.5, scale=2.5, size=num_matches)
    elif over <= 15:
        runs = np.random.normal(loc=10 + over*0.2, scale=3.0, size=num_matches)
    else:
        runs = np.random.normal(loc=12 + over*0.1, scale=4.0, size=num_matches)
    runs_data[over] = np.maximum(0, runs).astype(int) # Runs cannot be negative

# Prepare data for box plot - list of arrays, one array per over
plot_runs_data = [runs_data[over] for over in sorted(runs_data.keys())]
overs_labels = [str(over) for over in sorted(runs_data.keys())]

# --- Box Plot for Runs Scored Per Over ---
plt.figure(figsize=(10, 4))
plt.boxplot(plot_runs_data, tick_labels=overs_labels, patch_artist=True, medianprops={'color': 'red'})

plt.xlabel("Over Number", fontsize=12)
plt.ylabel("Runs Scored in Over", fontsize=12)
plt.title("Distribution of Runs Scored Per Over Across Multiple Matches", fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- Simulate Agricultural Data ---
np.random.seed(42) # for reproducibility

crop_types = ['Wheat', 'Rice', 'Corn', 'Soybeans']
years = np.arange(2010, 2024)

data = []
for year in years:
    for crop in crop_types:
        num_entries = 5 # Simulate multiple readings/farms per crop per year
        crop_area = np.random.normal(loc=100 + (len(crop_types) * 10), scale=20, size=num_entries)
        # Ensure area is positive
        crop_area = np.maximum(10, crop_area)

        # Production slightly varies with crop type and year
        base_production = np.random.normal(loc=500 + (len(crop) * 20) + ((year - 2010) * 5), scale=80, size=num_entries)
        production = np.maximum(50, base_production)

        productivity = production / crop_area # Productivity = Production / Area

        for i in range(num_entries):
            data.append({
                'Year': year,
                'Crop Type': crop,
                'Crop Area (acres)': crop_area[i],
                'Production (tons)': production[i],
                'Productivity (tons/acre)': productivity[i]
            })

df_agri = pd.DataFrame(data)

# --- Box Plot 1: Productivity Distribution by Year ---
plt.figure(figsize=(14, 7))
plt.boxplot([df_agri[df_agri['Year'] == year]['Productivity (tons/acre)'] for year in years],
            tick_labels=years, patch_artist=True, medianprops={'color': 'red'})
plt.xlabel("Year", fontsize=12)
plt.ylabel("Productivity (tons/acre)", fontsize=12)
plt.title("Distribution of Crop Productivity by Year", fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# --- Box Plot 2: Productivity Distribution by Crop Type ---
plt.figure(figsize=(10, 7))
plt.boxplot([df_agri[df_agri['Crop Type'] == crop]['Productivity (tons/acre)'] for crop in crop_types],
            tick_labels=crop_types, patch_artist=True, medianprops={'color': 'red'})
plt.xlabel("Crop Type", fontsize=12)
plt.ylabel("Productivity (tons/acre)", fontsize=12)
plt.title("Distribution of Crop Productivity by Crop Type", fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Agricultural box plots generated successfully!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Ensure df_agri is available from previous execution or define it for standalone execution
# If df_agri is not defined, run the agricultural box plot cell first.
# For demonstration, I'll include a minimal re-creation of df_agri here if it's not present.

# --- Minimal re-creation of df_agri for standalone execution (if needed) ---
# This block can be removed if df_agri is guaranteed to be in the kernel state.
if 'df_agri' not in locals():
    np.random.seed(42)
    crop_types = ['Wheat', 'Rice', 'Corn', 'Soybeans']
    years = np.arange(2010, 2013) # Reduced years for brevity in this example
    data = []
    for year in years:
        for crop in crop_types:
            num_entries = 2
            crop_area = np.random.normal(loc=100, scale=10, size=num_entries)
            production = np.random.normal(loc=500, scale=50, size=num_entries)
            productivity = production / crop_area
            for i in range(num_entries):
                data.append({
                    'Year': year,
                    'Crop Type': crop,
                    'Crop Area (acres)': crop_area[i],
                    'Production (tons)': production[i],
                    'Productivity (tons/acre)': productivity[i]
                })
    df_agri = pd.DataFrame(data)

# --- Prepare data for heatmap ---
# Aggregate the data to get average productivity per crop type per year
heatmap_data = df_agri.pivot_table(
    index='Crop Type',
    columns='Year',
    values='Productivity (tons/acre)',
    aggfunc='mean'
)

# --- Generate the Heatmap ---
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data,
    annot=True,     # Show the numerical values on the heatmap
    fmt=".2f",      # Format annotations to 2 decimal places
    cmap="Greens",  # Color map for agricultural productivity
    linewidths=.5,  # Add lines between cells
    linecolor='black'
)

plt.title('Average Crop Productivity (tons/acre) by Crop Type and Year', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Crop Type', fontsize=12)
plt.tight_layout()
plt.show()

print("Agricultural heatmap generated successfully!")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for 20 overs in a T20 match
np.random.seed(42) # for reproducibility
overs = np.arange(1, 21)
runs_per_over = np.random.randint(low=2, high=25, size=20)

# Find the index of the over with maximum and minimum runs
max_runs_over_idx = np.argmax(runs_per_over)
min_runs_over_idx = np.argmin(runs_per_over)

# Create a list of colors for the bars
# Default color will be blue, max run over green, min run over red
bar_colors = ['skyblue'] * len(overs)
bar_colors[max_runs_over_idx] = 'green'
bar_colors[min_runs_over_idx] = 'red'

# Plotting the bar graph
plt.figure(figsize=(10, 5))
plt.barh(overs, runs_per_over, color=bar_colors)

# Add labels and title
plt.xlabel("Over Number", fontsize=12)
plt.ylabel("Runs Scored", fontsize=12)
plt.title("Runs Scored Per Over in a T20 Match (Max/Min Highlighted)", fontsize=14, fontweight='bold')

# Add legend for colors (optional, but good for clarity)
plt.legend(handles=[
    plt.Rectangle((0,0),1,1, color='green', label=f'Max Runs ({runs_per_over[max_runs_over_idx]} in Over {overs[max_runs_over_idx]})'),
    plt.Rectangle((0,0),1,1, color='red', label=f'Min Runs ({runs_per_over[min_runs_over_idx]} in Over {overs[min_runs_over_idx]})'),
    plt.Rectangle((0,0),1,1, color='skyblue', label='Other Overs')
])

plt.xticks(overs) # Ensure all over numbers are displayed on x-axis
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data based on shot analysis
regions = ['Behind Square (Fine Leg/Third Man)', 'V-Region (Straight/Long-off/Long-on)',
           'Square of Wicket (Point/Cover)', 'Mid-Wicket']
# Estimated run distribution across 33 runs
runs_per_region = [12, 10, 6, 5]

# Visual customization
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0)  # Explode the signature "Behind Square" region

plt.figure(figsize=(10, 7))
plt.pie(runs_per_region, labels=regions, autopct='%1.1f%%', startangle=140,
        colors=colors, explode=explode, shadow=True)

plt.title('Suryakumar Yadav Shot Analysis vs Zimbabwe (33 off 13)\nICC T20 World Cup - Feb 26, 2026',
          fontsize=14, fontweight='bold')
plt.axis('equal')

plt.show()
```
