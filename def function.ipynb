```python
a = [10, 20, 30, 40, 50]
b = [1, 2, 3, 4, 5]


subtracted = list()
for a, b in zip(a, b):
    item = a - b
    subtracted.append(item)

print(subtracted)
```

```python
def REVERSE(a):
    a.reverse()
    return(a)

def YKNJS(a):
    b = []
    b.extend(REVERSE(a))
    print(b)

a = [1, 3.1, 5.31, 7.531]
YKNJS(a)
```

```python
li = [1, 1.33, 'GFG', 0, 'NO', None, 'G', True]
val1, val2 = 0, ' '
for x in li:
    if(type(x) == int or type(x) == float):
        val1 += x
    elif(type(x) == str):
        val2 += x
    else:
        break
print(val1, val2)
```
