```python
# String Manipulation in Python

## Creating Strings
"""Strings represent textual data such as crop names, soil types,
or farmer details."""

crop = "Rice"
soil_type = "Black Soil"
print(crop)
print(soil_type)

#Length of a String
#The len() function returns the number of characters in a string.

print(len(crop))
print(len(soil_type))
```

```python
# Indexing and Slicing
# Indexing accesses individual characters,
# while slicing extracts substrings.

crop_name = "Groundnut"
print(crop_name[0])      # First character
print(crop_name[-1])     # Last character
print(crop_name[:5])     # First five characters
print(crop_name[5:])     # Remaining characters
```

```python
#Case Conversion Methods
#Used to standardize textual agricultural data.

print("rice".upper())
print("WHEAT".lower())
print("green gram".title())
print("cotton crop".capitalize())
```

```python
#Trimming and Cleaning Strings
#Useful for cleaning user input and CSV data.

raw_crop = "   Rice   "
print(raw_crop.strip())
print("***Wheat".lstrip("*"))
print("Maize###".rstrip("#"))
```

```python
#Searching and Counting Substrings
#Helps in keyword-based filtering of crop records.

text = "Organic Rice and Rice cultivation"
print(text.find("Rice"))
print(text.count("Rice"))
print("Rice" in text)


#Validation and Checking Methods
#Used for validating form inputs.

print("Rice".isalpha())
print("2026".isdigit())
print("Plot12".isalnum())
print("   ".isspace())
```

```python
#Replacing String Content
#Used to update or correct crop records.

crop_info = "Rice Crop"
updated_crop_info = crop_info.replace("Rice", "Wheat")
print(updated_crop_info)


#Splitting and Joining Strings
#Common in CSV-style agriculture datasets.

record = "Rice,1200,Black Soil"
fields = record.split(",")
print(fields)

joined_record = ",".join(fields)
print(joined_record)

#String Formatting
#Formatting improves readability of output.

crop = "Rice"
yield_kg = 1200
print(f"Crop: {crop}, Yield: {yield_kg} kg")
print("Crop: {}, Yield: {} kg".format(crop, yield_kg))
```

```python
#Escape Characters
#Used for formatting text output.

print("Farmer's crop record")
print("Line1\nLine2")
print(sorted("wheaT"))
```

```python
crop_code = "CRP_RICE_2026"
print(crop_code.startswith("CRP"))
print(crop_code.endswith("2025"))
```

```python
#Extracting ALL strings with a prefix

codes = [
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
filtered = []

for crop_code in codes:
    if crop_code.startswith("CRP_"):#True
        filtered.append(crop_code)

print(filtered)
```

```python
#Extract and clean in one go

codes = [
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
clean_codes = []

for code in codes:
    if code.startswith("CRP_"):
        clean_codes.append(code[4:])

print(clean_codes)
```

```python
#List comprehension

clean_codes = [print(code[4:]) for code in codes if code.startswith("CRP_")]
```

```python
#split method

codes = [
    "CRP_RICE_2026",
    "CRP_WHEAT_2025",
    "FRT_MANGO_2026",
    "CRP_MAIZE_2024"
]
for crop_code in codes:
  if crop_code.startswith("CRP_"):
    code,crop,year = crop_code.split("_")
    print(crop, year)
```

```python
#Mini Agriculture Data Cleaning Example
#A consolidated example using multiple string operations.

raw_record = "  Rice,1200,Black Soil** "
clean_record = raw_record.strip()
clean_record1=clean_record.rstrip("*")
crop,yield_kg,soil = clean_record1.split(",")
crop = crop.title()

print(f"Crop: {crop}")
print(f"Yield: {yield_kg} kg")
print(f"Soil Type: {soil}")
```
