```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Farm_ID': ['F001', 'F002', 'F003', 'F004', 'F005', 'F006', 'F007', 'F008'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Crop':   ['Wheat', 'Rice', 'Corn', 'Wheat', 'Rice', 'Corn', 'Wheat', 'Rice'],
    'Yield_tons':    [180, 228, 416, 164, 245, 390, 195, 212],
    'Area_ha':        [45, 38, 52, 41, 40, 48, 48, 35],
    'Quality_Score': [8.5, 9.2, 8.8, 7.9, 9.0, 8.6, 8.3, 9.1]
}

df = pd.DataFrame(data)
df['Yield_per_ha'] = (df['Yield_tons'] / df['Area_ha']).round(2)

print('=== SORTED BY YIELD PER HECTARE ===')
sorted_df = df.sort_values('Yield_per_ha', ascending=False)
print(sorted_df[['Farm_ID', 'Crop', 'Yield_per_ha', 'Quality_Score']])

print('\n=== CROP DISTRIBUTION ===')
print(df['Crop'].value_counts())

print('\n=== REGIONAL DISTRIBUTION ===')
print(df['Region'].value_counts())

print('\n=== CROP PERCENTAGES ===')
print(df['Crop'].value_counts(normalize=True).mul(100).round(1))

# Ranking
df['Yield_Rank'] = df['Yield_per_ha'].rank(ascending=False)
df['Quality_Rank'] = df['Quality_Score'].rank(ascending=False)

print('\n=== TOP 3 FARMS BY YIELD ===')
top_yield = df.nsmallest(3, 'Yield_Rank')
print(top_yield[['Farm_ID', 'Crop', 'Yield_per_ha', 'Yield_Rank']])

print('\n=== UNIQUE CROPS ===')
print(f'Crops grown: {df["Crop"].unique()}')
print(f'Number of different crops: {df["Crop"].nunique()}')

# Correlation analysis
print('\n=== CORRELATION ANALYSIS ===')
numeric_cols = df[['Yield_per_ha', 'Quality_Score', 'Area_ha']]
print(numeric_cols.corr().round(3))

# Note: Actual plotting would create visualizations
# df.plot(kind='bar', x='Farm_ID', y='Yield_per_ha', title='Yield per Hectare by Farm')
# plt.ylabel('Tons per Hectare')
# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()
```

```python
import pandas as pd
import numpy as np

# Agricultural data with missing values
data = {
    'Farm_ID': ['F001', 'F002', 'F003', 'F004', 'F005', 'F006'],
    'Crop': ['Wheat', 'Rice', np.nan, 'Wheat', 'Corn', 'Rice'],
    'Yield_tons': [180, np.nan, 416, 164, np.nan, 228],
    'Rainfall_mm': [450, 820, np.nan, 420, 710, 850],
    'Fertilizer_kg': [120, 150, 200, np.nan, 195, 160],
    'Soil_pH': [6.5, 5.8, 6.2, 6.7, np.nan, 5.9]
}

df = pd.DataFrame(data)

print('Original Data with Missing Values:')
print(df)

print('\n=== MISSING VALUE ANALYSIS ===')
print('\nMissing values per column:')
print(df.isna().sum())
print(f'\nTotal missing values: {df.isna().sum().sum()}')
print(f'Percentage of missing data: {(df.isna().sum().sum() / df.size * 100):.2f}%')

print('\n=== ROWS WITH MISSING DATA ===')
rows_with_missing = df[df.isna().any(axis=1)]
print(rows_with_missing)

# Strategy 1: Drop rows with missing values
print('\n=== STRATEGY 1: DROP MISSING ROWS ===')
df_dropped = df.dropna()
print(f'Original shape: {df.shape}')
print(f'After dropping: {df_dropped.shape}')
print(df_dropped)

# Strategy 2: Fill with mean/median
print('\n=== STRATEGY 2: FILL WITH STATISTICS ===')
df_filled = df.copy()
# Fill numeric columns with mean
df_filled['Yield_tons'].fillna(df_filled['Yield_tons'].mean(), inplace=True)
df_filled['Rainfall_mm'].fillna(df_filled['Rainfall_mm'].median(), inplace=True)
df_filled['Fertilizer_kg'].fillna(df_filled['Fertilizer_kg'].mean(), inplace=True)
df_filled['Soil_pH'].fillna(df_filled['Soil_pH'].mean(), inplace=True)
# Fill categorical with mode
df_filled['Crop'].fillna(df_filled['Crop'].mode()[0], inplace=True)
print(df_filled)

# Strategy 3: Forward fill (useful for time series)
print('\n=== STRATEGY 3: FORWARD FILL ===')
df_ffill = df.fillna(method='ffill')
print(df_ffill)

# Strategy 4: Custom fill per column
print('\n=== STRATEGY 4: CUSTOM FILL ===')
df_custom = df.fillna({
    'Crop': 'Unknown',
    'Yield_tons': df['Yield_tons'].median(),
    'Rainfall_mm': 650, # Regional average
    'Fertilizer_kg': 150, # Standard recommendation
    'Soil_pH': 6.3 # Neutral pH
})
print(df_custom)

print('\n=== VERIFICATION ===')
print(f'Missing values after treatment: {df_custom.isna().sum().sum()}')
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data.csv')

print('Original Data with Missing Values:')
print(df)

print('\n=== MISSING VALUE ANALYSIS ===')
print('\nMissing values per column:')
print(df.isna().sum())
print(f'\nTotal missing values: {df.isna().sum().sum()}')
print(f'Percentage of missing data: {(df.isna().sum().sum() / df.size * 100):.2f}%')

print('\n=== ROWS WITH MISSING DATA ===')
rows_with_missing = df[df.isna().any(axis=1)]
print(rows_with_missing)

# Strategy 1: Drop rows with missing values
print('\n=== STRATEGY 1: DROP MISSING ROWS ===')
df_dropped = df.dropna()
print(f'Original shape: {df.shape}')
print(f'After dropping: {df_dropped.shape}')
print(df_dropped)

# Strategy 2: Fill with mean
print('\n=== STRATEGY 2: FILL WITH MEAN ===')
df_filled_mean = df.copy()
df_filled_mean['Calories'] = df_filled_mean['Calories'].fillna(df_filled_mean['Calories'].mean())
print(df_filled_mean)

# Strategy 3: Forward fill
print('\n=== STRATEGY 3: FORWARD FILL ===')
df_ffill = df.ffill()
print(df_ffill)

# Strategy 4: Fill with median
print('\n=== STRATEGY 4: FILL WITH MEDIAN ===')
df_filled_median = df.copy()
df_filled_median['Calories'] = df_filled_median['Calories'].fillna(df_filled_median['Calories'].median())
print(df_filled_median)

print('\n=== VERIFICATION (After median fill) ===')
print(f'Missing values after median fill: {df_filled_median.isna().sum().sum()}')
```

```python
import pandas as pd

# Multi-year farm data
data = {
    'Year': [2020, 2020, 2020, 2021, 2021, 2021, 2022, 2022, 2022, 2023, 2023, 2023],
    'Region': ['North', 'South', 'East', 'North', 'South', 'East',
               'North', 'South', 'East', 'North', 'South', 'East'],
    'Crop': ['Wheat', 'Rice', 'Corn', 'Wheat', 'Rice', 'Corn',
             'Wheat', 'Rice', 'Corn', 'Wheat', 'Rice', 'Corn'],
    'Yield_tons': [180, 228, 416, 195, 245, 430, 210, 250, 445, 225, 260, 455],
    'Area_ha': [50, 45, 55, 52, 46, 56, 54, 48, 57, 55, 49, 58],
    'Cost_thousands': [85, 95, 180, 90, 100, 190, 95, 105, 195, 100, 110, 200]
}

df = pd.DataFrame(data)

print('Agricultural Dataset:')
print(df)

# Group by Region
print('\n=== ANALYSIS BY REGION ===')
region_stats = df.groupby('Region').agg({
    'Yield_tons': ['sum', 'mean', 'std'],
    'Area_ha': 'sum',
    'Cost_thousands': 'sum'
})
print(region_stats)

# Group by Crop
print('\n=== ANALYSIS BY CROP ===')
crop_stats = df.groupby('Crop').agg({
    'Yield_tons': ['mean', 'min', 'max'],
    'Cost_thousands': 'mean'
}).round(2)
print(crop_stats)

# Multiple grouping (Region and Crop)
print('\n=== REGION-CROP ANALYSIS ===')
region_crop = df.groupby(['Region', 'Crop'])['Yield_tons'].mean().round(2)
print(region_crop)

# Year-over-year growth
print('\n=== YEARLY TRENDS ===')
yearly = df.groupby('Year').agg({
    'Yield_tons': 'sum',
    'Area_ha': 'sum'
})
yearly['Yield_per_ha'] = (yearly['Yield_tons'] / yearly['Area_ha']).round(2)
print(yearly)

# Transform: Add group mean to each row
df['Region_Avg_Yield'] = df.groupby('Region')['Yield_tons'].transform('mean').round(2)
df['Crop_Avg_Yield'] = df.groupby('Crop')['Yield_tons'].transform('mean').round(2)

print('\n=== DATASET WITH GROUP AVERAGES ===')
print(df[['Year', 'Region', 'Crop', 'Yield_tons',
          'Region_Avg_Yield', 'Crop_Avg_Yield']].head(6))

# Pivot table
print('\n=== PIVOT TABLE: Average Yield by Region and Crop ===')
pivot = df.pivot_table(values='Yield_tons',
                       index='Region',
                       columns='Crop',
                       aggfunc='mean').round(2)
print(pivot)

# Filter groups with high performance
print('\n=== HIGH PERFORMING REGIONS (Avg Yield > 300) ===')
high_regions = df.groupby('Region').filter(lambda x: x['Yield_tons'].mean() > 300)
print(high_regions[['Region', 'Crop', 'Yield_tons']].drop_duplicates('Region'))
```

```python
import pandas as pd

df = pd.read_csv('data.csv')

print('Original Data (first 5 rows):')
print(df.head())

# Group by Duration (similar to 'Analysis by Region')
print('\n=== ANALYSIS BY DURATION ===')
duration_stats = df.groupby('Duration').agg({
    'Pulse': ['sum', 'mean', 'std'],
    'Maxpulse': ['sum', 'mean', 'std'],
    'Calories': ['sum', 'mean', 'std']
}).round(2)
print(duration_stats)

# General trends for Calories by Duration (similar to 'Analysis by Crop')
print('\n=== CALORIES TRENDS BY DURATION ===')
calories_trends = df.groupby('Duration')['Calories'].agg(['mean', 'min', 'max']).round(2)
print(calories_trends)

# Transform: Add group mean of Calories to each row
df['Duration_Avg_Calories'] = df.groupby('Duration')['Calories'].transform('mean').round(2)

print('\n=== DATASET WITH DURATION AVERAGE CALORIES (first 6 rows) ===')
print(df[['Duration', 'Pulse', 'Maxpulse', 'Calories', 'Duration_Avg_Calories']].head(6))

# Filter groups with high average Calories
print('\n=== HIGH CALORIE DURATION GROUPS (Avg Calories > 350) ===')
high_calorie_durations = df.groupby('Duration').filter(lambda x: x['Calories'].mean() > 350)
print(high_calorie_durations[['Duration', 'Calories', 'Duration_Avg_Calories']].drop_duplicates('Duration'))
```
